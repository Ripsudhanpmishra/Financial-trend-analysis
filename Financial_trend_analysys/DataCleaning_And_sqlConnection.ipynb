```python
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
```

```python
data = pd.read_csv("data.csv", encoding="latin1")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.dropna(subset=['CustomerID'], inplace=True)
print(f"Shape after dropping CustomerID nulls: {data.shape}")
```

```python
data.isnull().sum()
```

```python
data = data[data['Quantity'] > 0]
print(f"Shape after removing non-positive quantities: {data.shape}")
```

```python
data.info()
```

```python
data.describe()
```

```python
# Convert 'InvoiceDate' to datetime objects if not already
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

print(f"Shape before filtering for 2010: {data.shape}")

# Filter the dataset for the year 2010
data_2010 = data[data['InvoiceDate'].dt.year == 2010]

print(f"Shape after filtering for 2010: {data_2010.shape}")
display(data_2010.head(12))
```

```python
data_2010.tail()
```

```python
data_2010.to_csv("data_2010.csv", index=False, encoding = 'utf-8')
```

```python
#connecting to sql server
engine = create_engine('mysql+mysqlconnector://root:123456@127.0.0.1/finance_db')
print("databse connected")
```

```python
data_2010.to_sql('transections', con=engine, if_exists='replace', index=False)
print(f"DataFrame 'data_2010' successfully loaded into table 'transections'.")
```

```python
query = "SELECT * FROM transections"
df = pd.read_sql(query, engine)
print(df.head())
```

```python
query1 = "SELECT YEAR(InvoiceDate) AS Year, QUARTER(InvoiceDate) AS Quarter, SUM(Quantity * UnitPrice) AS Total_Revenue, COUNT(DISTINCT InvoiceNo) AS Transaction_Count FROM transections GROUP BY Year, Quarter ORDER BY Year, Quarter"
query2 = "SELECT COUNT(DISTINCT q1.CustomerID) AS Churned_Customers FROM transections q1 LEFT JOIN transections q2 ON q1.CustomerID = q2.CustomerID AND QUARTER(q2.InvoiceDate) = 2 WHERE QUARTER(q1.InvoiceDate) = 1 AND q2.CustomerID IS NULL;"
```

```python
df2 = pd.read_sql(query1, engine)
print(df2.head())
```
